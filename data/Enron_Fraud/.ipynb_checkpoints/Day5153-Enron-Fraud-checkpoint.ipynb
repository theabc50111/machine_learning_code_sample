## 讀取資料

```python
train_data = pd.read_csv('train_data.csv')
test_features = pd.read_csv('test_features.csv')
sample_submission = pd.read_csv('sample_submission.csv')
print('train_data shape: ', train_data.shape)
print('test_features: ', test_features.shape)
```

```python
train_y = train_data['poi'].map({True: 1, False: 0}) # 'poi' 是預測結果
train_data = train_data.drop('poi',axis = 1)
name = test_features['name']
```

```python
alldata =  pd.concat([train_data,test_features])
train_num = train_data.shape[0]  # 113
alldata.head()
```

## 處理缺失值, 離群值

```python
alldata = alldata.drop(['name','email_address'],axis = 1) # 去除非數值欄位 & 預測結果
```

## 特徵選擇

```python
#alldata = alldata.drop(['loan_advances'],axis = 1)
alldata.shape
```

```python
for i in alldata.columns:
    alldata[i] = alldata[i].fillna(0)
```

```python
mms = MinMaxScaler()
for i in alldata.columns:
    alldata[i] = mms.fit_transform(alldata[i].values[:,np.newaxis])
alldata.describe()
```

```python
# 'loan_advances' 欄位做獨熱編碼
a = pd.get_dummies(alldata.loc[:,'loan_advances'])
alldata = pd.concat([alldata,a],axis = 1)
alldata.head()
```

```python
alldata = alldata.drop('loan_advances',axis = 1)
alldata.to_csv('r.csv')
```

```python
# 選擇模型
#clf = GradientBoostingClassifier()
clf = LogisticRegression()
```

## 交叉驗證

```python
# 羅列多種可能超參數 (超參數因模型而異 )

# GradientBoostingClassifier
# learning_rate = [0.07, 0.08,0.1,0.2]
# max_depth = [3, 4]
# subsample = [0.8,0.9]

# LogisticRegression
penalty = ['l1','l2']
class_weight = [None,'balanced']
max_iter = [100,150,200,250]


# 編成字典
#param_grid = dict(learning_rate = learning_rate, max_depth = max_depth, subsample = subsample)
param_grid = dict(penalty = penalty, class_weight = class_weight, max_iter = max_iter)

# 建立搜尋物件，放入模型及參數組合字典 (n_jobs=-1 會使用全部 cpu 平行運算; cv是 cross validation, 表示資料切割的份數 )
grid_search = GridSearchCV(clf, param_grid, scoring='roc_auc', n_jobs=-1, verbose=1,cv = 5,refit=True)

# 搜尋最佳參數
grid_result = grid_search.fit(alldata[:train_num], train_y)

# 查看最佳參數
print (grid_result.best_params_)

# 套用最佳參數至步驟 (3.0)
```

```python
clf = GradientBoostingClassifier(learning_rate= 0.08, max_depth= 3)
clf.fit(alldata[:train_num], train_y)
```

```python
clf = LogisticRegression(class_weight = 'balanced', max_iter=100)
clf.fit(alldata[:train_num], train_y)
```

```python
roc_auc_score(train_y,clf.predict(alldata[:train_num]))
```

```python
confusion_matrix(train_y,clf.predict(alldata[:train_num]))
```

## 製作 submission

```python
submission = pd.DataFrame()
submission['name'] = name
submission['poi'] = clf.predict(alldata[train_num:])
submission
submission.to_csv('submission_no_normalization.csv', index = False)
```
