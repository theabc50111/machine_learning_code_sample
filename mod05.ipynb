# EDA整體流程

#### 讀取檔案

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

data_path = ['./data/house_train.csv','./data/house_test.csv']
df_train = pd.read_csv(data_path[0])
df_test = pd.read_csv(data_path[1])

# 將答案抽離
df_train_Y = df_train['單價(元/平方公尺)']
df_train = df_train.drop(['單價(元/平方公尺)'] , axis=1)

# 記得刪index
df_concat = pd.concat([df_train,df_test],ignore_index = True)
df_concat = df_concat.drop(['index'] , axis=1)
display(df_concat.tail())
```

#### 格式調整

```python
print(df_concat.info())
print('-'*100)
print(df_concat.describe())
```

```python
# 只取 int64, float64 兩種數值型欄位, 存於 num_features 中, 其他類別存於notnum_features
num_features = []
notnum_features = []
for dtype, feature in zip(df_concat.dtypes, df_concat.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
    else:
        notnum_features.append(feature)

print(f'lehgth of all featrues : {len(df_concat.columns)}\n')
print(f'length of Numeric Features : {len(num_features)}\n Numeric Features : {num_features}\n')
print(f'length of Not Numeric Features : {len(notnum_features)}\n Not Numeric Features : {notnum_features}')
```

#### 偵測與填補遺失值

```python
df_concat_pre = pd.DataFrame()
for feature in num_features:
    if df_concat[feature].isnull().any():
        print(f'{feature} has missing value')
        df_concat_pre[feature] = df_concat[feature].fillna(df_concat[feature].mean())
    else:
        df_concat_pre[feature] = df_concat[feature]
display(df_concat_pre.head())
```

#### 特徵選擇

#### 文字類別資料轉數字

```python
import re 

df_concat_pre['交易筆棟數_土地']= df_concat['交易筆棟數'].apply(lambda x:int(re.findall(r'\d+',x)[0]))
df_concat_pre['交易筆棟數_建物']= df_concat['交易筆棟數'].apply(lambda x:int(re.findall(r'\d+',x)[1]))
df_concat_pre['交易筆棟數_車位']= df_concat['交易筆棟數'].apply(lambda x:int(re.findall(r'\d+',x)[2]))
num_features.extend(['交易筆棟數_土地', '交易筆棟數_建物', '交易筆棟數_車位']) ; notnum_features.remove('交易筆棟數')

display(df_concat_pre.loc[:,['交易筆棟數_土地','交易筆棟數_建物','交易筆棟數_車位']])
display(df_concat['交易筆棟數'])
```

```python
df_concat_pre['總樓層數'] = df_concat['總樓層數']
cn_in = {np.nan:np.nan, '四層':4, '二十一層':21, '十二層':12, '五層':5, '三層':3, '十三層':13, 
         '十四層':14, '六層':6, '十層':10,'九層':9, '八層':8, '七層':7, '十九層':19, '十五層':15, 
         '十一層':11, '十七層':17, '二層':2, '二十層':20, '十六層':16,'二十七層':27, '二十三層':23, 
         '十八層':18, '三十三層':33, '二十四層':24, '二十六層':26, '一層':1, '二十二層':20, 
         '四十二層':42, '二十五層':25, '三十五層':35, '二十九層':29, '二十八層':28, '三十層':30, 
         '三十八層':38, '三十一層':31}
df_concat_pre['總樓層數'] = df_concat_pre['總樓層數'].apply(lambda x : cn_in[x])
df_concat_pre['總樓層數'] = df_concat_pre['總樓層數'].fillna(int(df_concat_pre['總樓層數'].mean()))
num_features.append('總樓層數') ; notnum_features.remove('總樓層數')

print(df_concat_pre['總樓層數'].unique())
```

```python
df_concat_pre['low_use_electricity'] = df_concat['low_use_electricity'] 
df_concat_pre['low_use_electricity'][df_concat_pre['low_use_electricity'].isnull()] = 'empty' 
df_concat_pre['low_use_electricity'] = \
df_concat_pre['low_use_electricity'].apply(lambda x:x.replace('%',''))
df_concat_pre['low_use_electricity'][df_concat_pre['low_use_electricity']=='empty'] = np.nan
df_concat_pre['low_use_electricity'] = \
df_concat_pre['low_use_electricity'].apply(float)
df_concat_pre['low_use_electricity'] = \
df_concat_pre['low_use_electricity'].fillna(df_concat_pre['low_use_electricity'].mean())
num_features.append('low_use_electricity') ; notnum_features.remove('low_use_electricity')
df_concat_pre['low_use_electricity']
```

#### 路段整理(文字類型)

```python
pattern = r'(.+路)(.+段)|(.+路)|(.+街)|(.+道)|(.+段)'
df_concat_pre['土地區段位置/建物區段門牌'] = df_concat['土地區段位置/建物區段門牌']
df_concat_pre['土地區段位置/建物區段門牌'] = \
df_concat_pre['土地區段位置/建物區段門牌'].apply(lambda x : re.sub(r'臺北市..區','',x))
df_concat_pre['土地區段位置/建物區段門牌'] = \
df_concat_pre['土地區段位置/建物區段門牌'].apply(lambda x : re.sub(r'\w+鄰','',x))
for i,location in enumerate(df_concat_pre['土地區段位置/建物區段門牌']):
    if re.search(pattern,location) == None:
        print(df_concat['土地區段位置/建物區段門牌'][i],i)        
        df_concat_pre = df_concat_pre.drop(i)
        df_concat = df_concat.drop(i)
        df_train_Y = df_train_Y.drop(i)

df_concat = df_concat.reset_index(drop=True) # drop後記得要重製index
df_concat_pre = df_concat_pre.reset_index(drop=True)
df_train_Y = df_train_Y.reset_index(drop=True) 
```

```python
df_concat_pre['土地區段位置/建物區段門牌'] = \
df_concat_pre['土地區段位置/建物區段門牌'].apply(lambda x : re.search(pattern,x).group())
df_concat_pre['土地區段位置/建物區段門牌']
```

# 處理離群值

```python
for i,f in enumerate(num_features):
    ori_series = df_concat_pre[f].copy()
    qt1 = ori_series.quantile(q=0.25)
    qt3 = ori_series.quantile(q=0.75)
    iqr = qt3-qt1
    ori_series[ori_series>(qt3 + 1.5*iqr)] = qt3 + 1.5*iqr
    ori_series[ori_series<(qt1 - 1.5*iqr)] = qt1 - 1.5*iqr
    #--------------------------------------------------------#
    plt.title(f) 
    plt.boxplot(ori_series)
    plt.show()
    
```

```python
plt.hist(df_concat_pre['交易筆棟數_建物'])
plt.show()
```

```python
df_concat_pre['車位總價(元)'] = np.log1p(df_concat_pre['車位總價(元)'] )
df_concat_pre['車位移轉總面積(平方公尺)'] = np.log1p(df_concat_pre['車位移轉總面積(平方公尺)'] )
df_concat_pre['交易筆棟數_建物'] = np.log1p(df_concat_pre['交易筆棟數_建物'])
trans = ['車位總價(元)','車位移轉總面積(平方公尺)','交易筆棟數_建物']
outliers = [f for f in num_features if f not in trans]
outliers
```

```python
for i,f in enumerate(outliers):
    ori_series = df_concat_pre[f]
    qt1 = ori_series.quantile(q=0.25)
    qt3 = ori_series.quantile(q=0.75)
    iqr = qt3-qt1
    ori_series[ori_series>(qt3 + 1.5*iqr)] = qt3 + 1.5*iqr
    ori_series[ori_series<(qt1 - 1.5*iqr)] = qt1 - 1.5*iqr
    df_concat_pre[f] = ori_series
    #--------------------------------------------------------#
    plt.title(f) 
    plt.boxplot(df_concat_pre[f])
    plt.show()
```

# Normalization

```python
from sklearn.preprocessing import StandardScaler

# split data
split_point = len(df_train_Y)
# loc include last one
df_train_norm = df_concat_pre.loc[:(split_point-1),num_features] # only num_feature need to be normalized
df_test_norm = df_concat_pre.loc[split_point:,num_features] # only num_feature need to be normalizedd
s = StandardScaler()
scaler = StandardScaler().fit(df_train_norm)
array_train_norm = scaler.transform(df_train_norm)
array_test_norm = scaler.transform(df_test_norm)
array_concat = np.concatenate([array_train_norm,array_test_norm])
df_concat_pre_num = pd.DataFrame(array_concat,columns=num_features)
df_concat_pre = pd.concat([df_concat_pre_num,df_concat_pre.loc[:,notnum_features]],axis=1)
df_concat_pre.head()
```
